```python
import pandas as pd
import ChooseFlie
```

```python
file_path = ChooseFlie.choose_file()
```

```python
df = pd.read_excel(file_path,dtype={
            '卡号':str,
            '账号':str})
df.head()
```

```python
from collections import defaultdict
my_dict = defaultdict(list)
for k, v in zip(df['账号'],df['卡号']):
    my_dict[k].append(v)
```

```python
my_dict
```

```python
data = pd.DataFrame(columns=['账号', '卡号'])
temp_ = [str(it) for it in my_dict.keys()]
data['账号'] = temp_
temp = []
for item in my_dict.values():
    item = [str(it) for it in item]
    temp.append(','.join(item))
data['卡号'] = temp
```

```python
data.to_excel('1.xlsx')
```
